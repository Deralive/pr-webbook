```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, IntSlider, FloatSlider
from scipy.special import comb  # 使用 scipy 的 comb 函数计算组合数
```

```python
# 设置中文字体
plt.rcParams['font.family'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

def binomial_distribution(n=10, p=0.5):
    k = np.arange(0, n + 1)
    # 使用 scipy.special.comb 计算组合数，支持向量化操作
    binomial_coeff = comb(n, k)
    probabilities = binomial_coeff * (p ** k) * ((1 - p) ** (n - k))
    
    plt.figure(figsize=(10, 6))
    plt.bar(k, probabilities,color='skyblue')
    plt.xlabel('成功次数 k')
    plt.ylabel('概率 P(X = k)')
    plt.title(f'二项分布 b(n={n}, p={p})')
    plt.grid(True, alpha=0.3)
    plt.show()

n_slider = IntSlider(min=1, max=50, step=1, value=10, description='n (试验次数)')
p_slider = FloatSlider(min=0.1, max=0.9, step=0.1, value=0.5, description='p (成功概率)')

interactive_plot = interactive(binomial_distribution, n=n_slider, p=p_slider)
interactive_plot
```
